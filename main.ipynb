```python
from PIL import Image
import numpy as np

def converttoimage(path: str) -> np.ndarray:
    imag = Image.open(path)
    imag = imag.convert('RGB')
    imag = imag.resize((224, 224))
    imag_array = np.array(imag)
    return imag_array
   
```
